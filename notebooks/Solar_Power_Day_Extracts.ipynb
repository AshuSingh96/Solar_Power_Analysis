```python
import pandas as pd
```

```python
df_plant1 = pd.read_csv('plant1_cleaned_data.csv')
df_plant2 = pd.read_csv('plant2_cleaned_data.csv')
```

```python
df_plant1
```

```python
# Convert the 'DATE_TIME' column to datetime format if it is not already
df_plant1['DATE_TIME'] = pd.to_datetime(df_plant1['DATE_TIME'])
df_plant2['DATE_TIME'] = pd.to_datetime(df_plant2['DATE_TIME'])

# Extract the day and create a new 'DAY' column
df_plant1['DAY'] = df_plant1['DATE_TIME'].dt.day
df_plant2['DAY'] = df_plant2['DATE_TIME'].dt.day

# Save back to CSV or proceed with further processing
df_plant1.to_csv("plant1_cleaned_data_with_day.csv", index=False)
df_plant2.to_csv("plant2_cleaned_data_with_day.csv", index=False)
```
